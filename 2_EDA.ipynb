### **<font color="#ff5653">1. 1_internet_100hogares_por_provincia.csv ###################################################</font>**

Graficos que se elaboraron:

- Acceso a internet por cada 100 hogares (Año/Accesos).
- Acceso a internet por cada 100 hogares (Trimestre/Accesos).

```python
# Cargamos el archivo.
df_1 = pd.read_csv("/content/drive/MyDrive/Datasets_ETL/1_internet_100hogares_por_provincia.csv")
```

```python
# Vista previa del dataframe.
df_1.head(5)
```

```python
# Grafico de dispersion y linea de tendencia: Acceso a internet por cada 100 hogares (Año/Accesos).

fig = plt.figure(figsize=(16, 8))                           # Tamaño del grafico
ax = fig.add_subplot(111)

x = df_1["Año"]                                             # Variable independiente: eje x
y = df_1["Accesos por cada 100 hogares"]                    # Variable dependiente: eje y

p = np.polyfit(x, y, 1)                                     # Genera la linea de tendencia

ax.scatter(x, y, color="#add5fa", s=10)                     # Circulos
ax.plot(x, np.polyval(p, x), color="#fa5f49")               # Linea

ax.set_title("Accesos a Internet por cada 100 hogares")     # Titulo del grafico
ax.set_xlabel("Año")                                        # Nombre eje x
ax.set_ylabel("Accesos")                                    # Nombre eje y

plt.show()                                                  # Muestra el grafico
```

Analisis:

- Se observa que desde el año 2014 al año 2022 hay una tendencia creciente en el acceso a internet en la Republica Argentina.

- Para realizar los KPI se compararan cuatrimestres de un año con respecto al año anterior para sacar el incremento de accesos a internet.

- Se observan algunos outliers, pero no lo podemos confirmar que realmente son outliers por falta de informacion sobre la toma de datos.

----------------------------------------------------------------------------------------------------------------------------------------------------------------

```python
# Grafico de dispersion y una linea de tendencia: Acceso a internet cada 100 hogares (Trimestre/Accesos).

fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111)

x = df_1["Trimestre"]
y = df_1["Accesos por cada 100 hogares"]

p = np.polyfit(x, y, 1)

ax.scatter(x, y, color="#add5fa", s=10)
ax.plot(x, np.polyval(p, x), color="#fa5f49")

ax.set_title("Accesos a Internet por cada 100 hogares")
ax.set_xlabel("Trimestre")
ax.set_ylabel("Accesos")

plt.show()
```

Analisis:

- Se observa una leve tendencia en el acceso a internet en los ultimos dos cuatrimestres del año.

### **<font color="#ff5653">2_internet_rangovelocidad_velocidadm_provincia.csv ############################################</font>**

Graficos que se elaboraron:

- Velocidad media de bajada de internet.
- Rangos de velocidad de bajada de internet.
- Velocidad de bajada "HASTA 512 kbps".
- Velocidad de bajada "+ 512 Kbps - 1 Mbps".
- Velocidad de bajada "+ 1 Mbps - 6 Mbps".
- Velocidad de bajada "+ 6 Mbps - 10 Mbps".
- Velocidad de bajada "+ 10 Mbps - 20 Mbps".
- Velocidad de bajada "+ 20 Mbps - 30 Mbps".
- Velocidad de bajada "+ 30 Mbps".
- Velocidad de bajada de internet para el año 2022.



```python
# Cargamos el archivo.
df_2 = pd.read_csv("/content/drive/MyDrive/Datasets_ETL/2_internet_rangovelocidad_velocidadm_provincia.csv")
```

```python
# Vista previa del dataframe.
df_2.head(2)
```

```python
# Grafico de dispersion y linea de tendencia: Velocidad media de bajada de internet.

fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111)

x = df_2["Año"]
y = df_2["velocidad_media_conexion(Mbps)"]

p = np.polyfit(x, y, 1)

ax.scatter(x, y, color="#add5fa", s=10)
ax.plot(x, np.polyval(p, x), color="#fa5f49")

ax.set_title("Velocidad media de bajada de internet")
ax.set_xlabel("Año")
ax.set_ylabel("Velocidad media (Mbps)")

plt.show()
```

Analisis:

- Se observa que desde el año 2014 la velocidad media de bajada de internet fue creciente, siendo en el año 2014 inferior a 5 Mbps llegando a superar en el año 2022 los 15 Mbps.

- Para realizar los KPI se compararan cuatrimestres de un año con respecto al año anterior para sacar el incremento de velocidad media de bajada de internet.

```python
# Grafico dispersion y lineas de tendencia: Rangos de velocidad de bajada de internet.
x = df_2["Año"]
y = df_2[["HASTA 512 kbps", "+ 512 Kbps - 1 Mbps", "+ 1 Mbps - 6 Mbps", "+ 6 Mbps - 10 Mbps", "+ 10 Mbps - 20 Mbps", "+ 20 Mbps - 30 Mbps", "+ 30 Mbps"]]

fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111)

for i, column in enumerate(y.columns):
    p = np.polyfit(x, y[column], 1)
    sns.regplot(x=x, y=y[column], ax=ax, color=sns.color_palette()[i], label=column)
    ax.plot(x, np.polyval(p, x), color=sns.color_palette()[i])

ax.set_title("Rangos de velocidad de bajada de internet")
ax.set_xlabel("Año")
ax.set_ylabel("Velocidad de bajada (Mbps)")
ax.legend()

plt.show()
```

Analisis:

- Podemos ver en en grafico en cuanto a los rangos de velocidades que el rango mas de 30 Mbps fue creciente cada año y que el rango de velocidad entre 1 a 6 Mbps fue decreciendo y que el resto de los rangos de velocidades se mantienen constantes durante los años analizados (2014 al 2022).

- A continuacion se muestran los graficos de los dintintos rangos de velocidades por separados desde el año 2014 al 2022 y se aprecia mejor las variaciones durante los años analizados.

```python
# Grafico de dispersion y una linea de tendencia: Velocidad de bajada "HASTA 512 kbps".

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

x = df_2["Año"]
y = df_2["HASTA 512 kbps"]

p = np.polyfit(x, y, 1)

ax.scatter(x, y, color="#87cefa", s=10)
ax.plot(x, np.polyval(p, x), color="#87cefa")

ax.set_title("Velocidad de bajada de internet hasta 512 kbps")
ax.set_xlabel("Año")
ax.set_ylabel("Velocidad media (Mbps)")

plt.show()
```

### **<font color="#ff5653">4. 5_internet_portecnologia_porprovincia.csv ##################################################</font>**

Graficos que se elaboraron:

- Tipo de conexion a internet por provincia para el año 2022.
- Evolución del tipo de conexión a internet por año (2014 a 2022).

```python
# Abrimos el archivo.
df_4 = pd.read_csv("/content/drive/MyDrive/Datasets_ETL/5_internet_portecnologia_porprovincia.csv")
```

```python
# Vista previa del dataframe.
df_4
```

```python
# Grafico de barras: Tipo de conexion a internet por provincia.

df_2022 = df_4[df_4["Año"] == 2022]                                                                            # Filtrar los datos para el año 2022.

df_grouped = df_2022.groupby("Provincia")[["ADSL", "Cablemodem", "Fibra óptica", "Wireless", "Otros"]].sum()   # Agrupar los datos por provincia y sumar las columnas de interes.

ax = df_grouped.plot(kind="bar", figsize=(18, 8), rot=0)                                                       # Grafico de barras.
plt.xticks(rotation=90)

ax.set_title("Tipo de conexión de internet por provincia en 2022")                                             # Titulos y etiquetas.
ax.set_xlabel("Provincia")
ax.set_ylabel("Cantidad")
plt.legend(loc="upper right")
plt.show()
```

Analisis:

- Si tomamos las provincias que mas se conectan a internet por Cablemodem son Cordoba, Santa Fe y Entre Rios.

- Las provincias que mas se conectan a internet por Fibra optica son Cordoba, Santa Fe y Tierra del Fuego.

- Las provincias que mas se conectan a internet por ADSL son Buenos Aires, Cordoba y Santa Fe.

```python
# Grafico lineal: Evolución del tipo de conexión a internet por año.

df_grouped = df_4.groupby("Año")[["ADSL", "Cablemodem", "Fibra óptica", "Wireless", "Otros"]].sum()

ax = df_grouped.plot(kind="line", figsize=(16, 8), rot=0)

ax.set_title("Evolución del tipo de conexión a internet por año")
ax.set_xlabel("Año")
ax.set_ylabel("Accesos")
plt.legend(loc="upper left")
plt.show()
```
